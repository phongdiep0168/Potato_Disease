```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
from IPython.display import HTML

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.config.experimental.list_physical_devices()
```

```python
IMAGE_SIZE = 256
CHANNELS = 3
train_datagen = ImageDataGenerator(
#     rescale = 1./255,   #Điều chỉnh kích thước
#     brightness_range = (0.5, 1.0), #Tăng độ sáng
#     zoom_range = 0.2,   #Phóng to, thu nho
#     vertical_flip = True,   #Xoay dọc
#     horizontal_flip=True,   #Xoay ngang
#     rotation_range = 10,    #xoay góc
)
train_generator =  train_datagen.flow_from_directory(
    'dataset/train',
    target_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
    #save_to_dir = "AugmentedImages"
)
```

```python
train_generator.class_indices
```

```python
class_names = list(train_generator.class_indices.keys())
class_names
```

```python
for image_batch,label_batch in train_generator:
    print(image_batch.shape)
    print(image_batch[0])#numpy array
    break

    
```

```python
validation_datagen = ImageDataGenerator(
#     rescale = 1./255,
#     horizontal_flip =True,
#     rotation_range = 10,
)
validation_generator =  validation_datagen.flow_from_directory(
    'dataset/val',
    target_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
)
```

```python
test_datagen = ImageDataGenerator(
#     rescale = 1./255,
#     horizontal_flip =True,
#     rotation_range = 10,
)
test_generator =  test_datagen.flow_from_directory(
    'dataset/test',
    target_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
)
```

```python
#CNN
input_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes = 3
#Set up layer
model = models.Sequential([
    layers.Conv2D(32,(3,3),activation = 'relu',input_shape = input_shape ),# 32 layer:Trial and Error to detect edges
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, kernel_size = (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, kernel_size = (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation = 'relu'),
    layers.Dense(n_classes,activation = 'softmax'),#n_classes:3 neuron,softmax : normalize the probability of your classes
])
```

```python
model.summary()
```

```python
#compile using optimizer "adam"
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    metrics = ['accuracy']
)
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch = 148,
    batch_size = 32,
    validation_data = validation_generator,
    validation_steps = 19,
    verbose = 1,
    epochs = 20,
)
```

```python
scores = model.evaluate(test_generator)
```
